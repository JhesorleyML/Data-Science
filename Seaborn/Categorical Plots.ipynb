```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tips, estimator=np.std, hue='sex')
```

```python
sns.countplot(x='sex',data=tips, hue='sex')
```

```python
sns.boxplot(x='day',y='total_bill', data=tips, hue='day')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips, jitter=True, hue='sex', dodge=True)
#split produces an error
```

```python
sns.violinplot(x='day',y='total_bill',data=tips, hue='day')
sns.swarmplot(x='day',y='total_bill',data=tips, color='black')
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='bar', hue='day') 
#factorplot is deprecated
```
