```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
#plot on the axes
axes1.plot(x,y)
axes1.set_title('Bigger Plot')

axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
# for current_ax in axes:
#     current_ax.plot(x,y)
axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')

plt.tight_layout()
```

# Figure Size and DPI

```python
fig,axes = plt.subplots(nrows=2, ncols=1, figsize=(8,6))

axes[0].plot(x, x**2, label='X Squared')
axes[0].plot(x,x**3, label='x Cubed')
axes[0].set_title('Figure 1')
axes[0].legend(loc=10)


axes[1].plot(x**2,x, label='Squared X')
axes[1].plot(x**3,x, label='Cubed X')
axes[1].set_title('Figure 2')
axes[1].legend()

plt.tight_layout()
```
